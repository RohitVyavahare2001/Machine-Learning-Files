```python
#convert target variable (categorical) to numerical

df["diagnosis_num"]=[1 if i=="M" else 0 for i in df["diagnosis"]]
```

```python
df.tail()
```

```python
#remove diagnosis column

df=df.drop(["diagnosis"],axis=1)
```

```python
df.shape
```

```python
df.head()
```

```python
# split data into feature set,target set

X=df.drop(["diagnosis_num"],axis=1)
y=df["diagnosis_num"]
```

```python
#split data into training set and testing set

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
print("X_train shape : ",X_train.shape)
print("X_test shape : ",X_test.shape)
print("y_train shape : ",y_train.shape)
print("y_test shape : ",y_test.shape)
```

```python
#model

nb=GaussianNB()
nb.fit(X_train,y_train)
```

```python
print("naive bayes score : ",nb.score(X_test,y_test))
```

```python
# 2nd version
```

```python
# scaling on X (feature set)

X_scaled=(X-np.min(X))/(np.max(X)-np.min(X))
```

```python
#split data into training and testing set

from sklearn.model_selection import train_test_split
X_train1,X_test1,y_train1,y_test1=train_test_split(X_scaled,y,test_size=0.3,random_state=42)
```

```python
#model

nb1=GaussianNB()
nb1.fit(X_train1,y_train1)
```

```python
print("naive bayes score after scaling : ",nb1.score(X_test1,y_test1))
```
